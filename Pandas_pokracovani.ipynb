```python
import pandas as pd
```

```python

planety = pd.DataFrame({
    "jmeno": ["Merkur", "Venuše", "Země", "Mars", "Jupiter", "Saturn", "Uran", "Neptun"],
    "symbol": ["☿", "♀", "⊕", "♂", "♃", "♄", "♅", "♆"],
    "obezna_poloosa": [0.39, 0.72, 1.00, 1.52, 5.20, 9.54, 19.22, 30.06],
    "obezna_doba": [0.24, 0.62, 1, 1.88, 11.86, 29.46, 84.01, 164.8],
})
planety = planety.set_index("jmeno")   
planety
```

```python
# přidání sloupce (Series), přiřadím jako hodnotu do slovníku:
mesice = [0, 0, 1, 2, 79, 82, 27, 14]
planety["mesice"] = mesice
planety
# toto je akceptovatelná výjimka, jinak zákon neměnit původní tabulku!
```

```python
# lepší je metoda assign - nemění tabulku, ale vytvoří kopii, jen je dočasný, nebo tabulku zkopírovat:
# planety2 = planety.copy()
# vytvořím další sloupec s rokem objeveno:
planety.assign(
    objevena=["starověk", "1600 BC", "neva", "3500 BC", "Galileo", "nevim", "hovno", "malej"]
)
planety2 = planety.copy()
planety2["objevena"] = ["starověk", "1600 BC", "neva", "3500 BC", "Galileo", "nevim", "hovno", "malej"]
planety2
```

```python
# přidání řádku = loc()

planety2.loc["Pluto"] = ["♇", 39.48, 247.94, 5, 2006]   # Seznam hodnot v řádku
planety2
```

```python
# přepis hodnoty:

planety2.loc["Pluto", "objevena"] = "v prdeli"
planety2
```

```python
# lze zapsat i do řádku a sloupce, které neexistují:
planety_bad = planety.copy()     # Pro jistotu si uděláme kopii

planety_bad.loc["Zeme", "planeta"] = True
planety_bad
```

```python
# přiřazení:
planety.loc["Merkur":"Mars", "je_obr"] = False
planety.loc["Jupiter":"Neptun", "je_obr"] = [True, True, True, True]
planety
```

```python
# když obr, tak prstenec:
planety.loc["Jupiter":"Neptun" , "ma_prstenec"] = True
planety.loc["Merkur":"Mars" , "ma_prstenec"] = False
planety
```
